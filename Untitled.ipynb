```python
# ============================================================
# 1. IMPORTS
# ============================================================

import os
import json
import math
import random
from dataclasses import dataclass, asdict
from itertools import combinations
from typing import Dict, List, Tuple

import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    set_seed,
)
```

```python
# ============================================================
# 2. CONFIG
# ============================================================

from dataclasses import dataclass, asdict
import os

@dataclass
class Config:
    # --------------------------
    # model
    # --------------------------
    model_name: str = "Qwen/Qwen2.5-1.5B-Instruct"

    # --------------------------
    # data paths
    # --------------------------
    story_file: str = "./gryph_writingprompts_sft_fine_story.jsonl"
    litbench_file: str = "./litbench_winner_only_story.jsonl"
    argugpt_file: str = "./argugpt_highscore_sft_logic.jsonl"
    infinity_eval_file: str = "./infinity_eval_queries.jsonl"

    # --------------------------
    # output
    # --------------------------
    output_dir: str = "./openended_transfer_exp_change_based"

    # --------------------------
    # mixing ratios
    # --------------------------
    story_ratio: float = 0.7
    litbench_ratio: float = 0.2
    argugpt_ratio: float = 0.1

    # --------------------------
    # split
    # --------------------------
    val_ratio: float = 0.05
    seed: int = 42

    # --------------------------
    # token length
    # --------------------------
    max_length: int = 512

    # --------------------------
    # training
    # --------------------------
    train_batch_size: int = 1
    eval_batch_size: int = 1
    learning_rate: float = 1e-5
    num_epochs: int = 1
    grad_accum_steps: int = 4
    weight_decay: float = 0.0
    max_grad_norm: float = 1.0

    # --------------------------
    # logging
    # --------------------------
    log_every: int = 20
    eval_every_loss_only: int = 200   # loss만 보는 가벼운 평가

    # --------------------------
    # selective FT
    # --------------------------
    top_neuron_ratio: float = 0.05

    # --------------------------
    # generation eval
    # --------------------------
    num_eval_queries: int = 20
    num_generations_per_query: int = 3
    gen_max_new_tokens: int = 64
    gen_temperature: float = 0.9
    gen_top_p: float = 0.95

    # --------------------------
    # precision
    # --------------------------
    use_bfloat16_if_available: bool = True


config = Config()
os.makedirs(config.output_dir, exist_ok=True)

print(asdict(config))

for p in [
    config.story_file,
    config.litbench_file,
    config.argugpt_file,
    config.infinity_eval_file,
]:
    print(p, "->", os.path.exists(p))
```

```python
# ============================================================
# 3. UTIL FUNCTIONS
# ============================================================

def ensure_dir(path: str):
    os.makedirs(path, exist_ok=True)

def save_json(obj, path: str):
    with open(path, "w", encoding="utf-8") as f:
        json.dump(obj, f, ensure_ascii=False, indent=2)

def read_jsonl(path: str) -> List[dict]:
    rows = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            rows.append(json.loads(line))
    return rows

def shuffle_list(lst: List[dict], seed: int):
    rnd = random.Random(seed)
    lst = lst[:]
    rnd.shuffle(lst)
    return lst

def get_dtype(cfg: Config):
    if torch.cuda.is_available():
        if cfg.use_bfloat16_if_available and torch.cuda.is_bf16_supported():
            return torch.bfloat16
        return torch.float16
    return torch.float32

set_seed(config.seed)
device = "cuda" if torch.cuda.is_available() else "cpu"
print("device:", device)
```

```python
# ============================================================
# 4. LOAD AND MIX TRAINING DATA
# ============================================================

def load_and_mix_training_data(cfg: Config) -> List[dict]:
    story_data = shuffle_list(read_jsonl(cfg.story_file), cfg.seed)
    litbench_data = shuffle_list(read_jsonl(cfg.litbench_file), cfg.seed + 1)
    argugpt_data = shuffle_list(read_jsonl(cfg.argugpt_file), cfg.seed + 2)

    total_target = len(story_data)

    n_story = int(total_target * cfg.story_ratio)
    n_litbench = int(total_target * cfg.litbench_ratio)
    n_argugpt = int(total_target * cfg.argugpt_ratio)

    mixed = (
        story_data[:min(n_story, len(story_data))] +
        litbench_data[:min(n_litbench, len(litbench_data))] +
        argugpt_data[:min(n_argugpt, len(argugpt_data))]
    )

    mixed = shuffle_list(mixed, cfg.seed + 999)
    return mixed

def split_train_val(data: List[dict], val_ratio: float, seed: int):
    data = shuffle_list(data, seed)
    n_val = max(1, int(len(data) * val_ratio))
    val_data = data[:n_val]
    train_data = data[n_val:]
    return train_data, val_data

mixed_data = load_and_mix_training_data(config)
train_rows, val_rows = split_train_val(mixed_data, config.val_ratio, config.seed)

print("total mixed:", len(mixed_data))
print("train:", len(train_rows))
print("val:", len(val_rows))
print("sample:", train_rows[0])
```

```python
# ============================================================
# 5. LOAD EVAL QUERIES
# ============================================================

def load_eval_queries(path: str, num_queries: int, seed: int) -> List[str]:
    rows = shuffle_list(read_jsonl(path), seed)
    rows = rows[:num_queries]

    queries = []
    for r in rows:
        if "instruction" in r:
            queries.append(r["instruction"])
        elif "query" in r:
            queries.append(r["query"])
    return queries

eval_queries = load_eval_queries(
    config.infinity_eval_file,
    config.num_eval_queries,
    config.seed,
)

print("num eval queries:", len(eval_queries))
print("sample query:", eval_queries[0])
```

```python
# ============================================================
# 6. DATASET AND COLLATOR
# ============================================================

class SFTDataset(Dataset):
    def __init__(self, rows: List[dict], tokenizer, max_length: int = 1024):
        self.rows = rows
        self.tokenizer = tokenizer
        self.max_length = max_length

    def format_example(self, instruction: str, response: str) -> str:
        return f"Instruction: {instruction}\nResponse: {response}"

    def __len__(self):
        return len(self.rows)

    def __getitem__(self, idx):
        row = self.rows[idx]
        text = self.format_example(row["instruction"], row["response"])
        enc = self.tokenizer(
            text,
            truncation=True,
            max_length=self.max_length,
            padding=False,
            return_tensors=None,
        )
        return {
            "input_ids": enc["input_ids"],
            "attention_mask": enc["attention_mask"],
            "labels": enc["input_ids"][:],
        }

class SFTCollator:
    def __init__(self, tokenizer):
        self.tokenizer = tokenizer

    def __call__(self, batch):
        input_ids = [x["input_ids"] for x in batch]
        attention_mask = [x["attention_mask"] for x in batch]
        labels = [x["labels"] for x in batch]

        padded_inputs = self.tokenizer.pad(
            {"input_ids": input_ids, "attention_mask": attention_mask},
            return_tensors="pt",
        )
        padded_labels = self.tokenizer.pad(
            {"input_ids": labels},
            return_tensors="pt",
        )["input_ids"]

        padded_labels[padded_labels == self.tokenizer.pad_token_id] = -100
        padded_inputs["labels"] = padded_labels
        return padded_inputs
```

```python
# ============================================================
# 7. BUILD MODEL/TOKENIZER AND DATALOADERS
# ============================================================

def build_model_and_tokenizer(cfg: Config, device: str):
    tokenizer = AutoTokenizer.from_pretrained(cfg.model_name, use_fast=True)
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token

    model = AutoModelForCausalLM.from_pretrained(
        cfg.model_name,
        dtype=get_dtype(cfg),
    )
    model.to(device)
    model.train()
    return model, tokenizer

tmp_tokenizer = AutoTokenizer.from_pretrained(config.model_name, use_fast=True)
if tmp_tokenizer.pad_token is None:
    tmp_tokenizer.pad_token = tmp_tokenizer.eos_token

train_dataset = SFTDataset(train_rows, tmp_tokenizer, max_length=config.max_length)
val_dataset = SFTDataset(val_rows, tmp_tokenizer, max_length=config.max_length)

train_loader = DataLoader(
    train_dataset,
    batch_size=config.train_batch_size,
    shuffle=True,
    collate_fn=SFTCollator(tmp_tokenizer),
)

val_loader = DataLoader(
    val_dataset,
    batch_size=config.eval_batch_size,
    shuffle=False,
    collate_fn=SFTCollator(tmp_tokenizer),
)

print("train batches:", len(train_loader))
print("val batches:", len(val_loader))
```

```python
# ============================================================
# 8. BASIC LOSS EVALUATION
# ============================================================

@torch.no_grad()
def evaluate_loss(model, dataloader, device):
    model.eval()
    total_loss = 0.0
    total_count = 0

    for batch in dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        bs = batch["input_ids"].size(0)
        total_loss += float(outputs.loss.item()) * bs
        total_count += bs

    avg_loss = total_loss / max(total_count, 1)
    ppl = math.exp(avg_loss) if avg_loss < 20 else float("inf")
    model.train()
    return {"val_loss": avg_loss, "val_ppl": ppl}
```

```python
# ============================================================
# 9. DIVERSITY + CONVERGENCE METRICS
# ============================================================

def get_ngrams(tokens: List[str], n: int):
    return [tuple(tokens[i:i+n]) for i in range(len(tokens)-n+1)]

def distinct_n(texts: List[str], n: int = 1) -> float:
    total = 0
    uniq = set()
    for text in texts:
        toks = text.split()
        ngrams = get_ngrams(toks, n)
        total += len(ngrams)
        uniq.update(ngrams)
    if total == 0:
        return 0.0
    return len(uniq) / total

def jaccard_similarity(a: str, b: str) -> float:
    sa = set(a.split())
    sb = set(b.split())
    union = sa | sb
    if len(union) == 0:
        return 0.0
    return len(sa & sb) / len(union)

def avg_pairwise_similarity(texts: List[str]) -> float:
    if len(texts) < 2:
        return 0.0
    scores = []
    for a, b in combinations(texts, 2):
        scores.append(jaccard_similarity(a, b))
    return float(np.mean(scores))

def avg_pairwise_diversity(texts: List[str]) -> float:
    return 1.0 - avg_pairwise_similarity(texts)

def cross_similarity(a_list: List[str], b_list: List[str]) -> float:
    scores = []
    for a in a_list:
        for b in b_list:
            scores.append(jaccard_similarity(a, b))
    if len(scores) == 0:
        return 0.0
    return float(np.mean(scores))
```

```python
# ============================================================
# 10. GENERATION FUNCTIONS
# ============================================================

@torch.no_grad()
def generate_multiple_responses(
    model,
    tokenizer,
    query: str,
    device: str,
    num_generations: int,
    max_new_tokens: int,
    temperature: float,
    top_p: float,
):
    model.eval()
    enc = tokenizer(query, return_tensors="pt").to(device)
    responses = []

    for _ in range(num_generations):
        out = model.generate(
            **enc,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            temperature=temperature,
            top_p=top_p,
            pad_token_id=tokenizer.eos_token_id,
        )
        text = tokenizer.decode(out[0], skip_special_tokens=True)
        if text.startswith(query):
            text = text[len(query):].strip()
        responses.append(text)

    model.train()
    return responses
```

```python
# ============================================================
# 11. OPEN-ENDED EVALUATION
# ============================================================

@torch.no_grad()
def evaluate_single_model_openended(
    model,
    tokenizer,
    queries: List[str],
    device: str,
    num_generations: int,
    max_new_tokens: int,
    temperature: float,
    top_p: float,
):
    all_results = []
    d1_scores = []
    d2_scores = []
    conv_scores = []
    div_scores = []

    for q in queries:
        responses = generate_multiple_responses(
            model=model,
            tokenizer=tokenizer,
            query=q,
            device=device,
            num_generations=num_generations,
            max_new_tokens=max_new_tokens,
            temperature=temperature,
            top_p=top_p,
        )

        d1 = distinct_n(responses, n=1)
        d2 = distinct_n(responses, n=2)
        conv = avg_pairwise_similarity(responses)
        div = avg_pairwise_diversity(responses)

        d1_scores.append(d1)
        d2_scores.append(d2)
        conv_scores.append(conv)
        div_scores.append(div)

        all_results.append({
            "query": q,
            "responses": responses,
            "distinct_1": d1,
            "distinct_2": d2,
            "intra_model_convergence": conv,
            "intra_model_diversity": div,
        })

    metrics = {
        "avg_distinct_1": float(np.mean(d1_scores)),
        "avg_distinct_2": float(np.mean(d2_scores)),
        "avg_intra_model_convergence": float(np.mean(conv_scores)),
        "avg_intra_model_diversity": float(np.mean(div_scores)),
    }
    return metrics, all_results


@torch.no_grad()
def evaluate_two_models_openended(
    model_a,
    model_b,
    tokenizer,
    queries: List[str],
    device: str,
    num_generations: int,
    max_new_tokens: int,
    temperature: float,
    top_p: float,
):
    all_results = []
    a_conv_scores = []
    b_conv_scores = []
    cross_scores = []

    for q in queries:
        resp_a = generate_multiple_responses(
            model=model_a,
            tokenizer=tokenizer,
            query=q,
            device=device,
            num_generations=num_generations,
            max_new_tokens=max_new_tokens,
            temperature=temperature,
            top_p=top_p,
        )
        resp_b = generate_multiple_responses(
            model=model_b,
            tokenizer=tokenizer,
            query=q,
            device=device,
            num_generations=num_generations,
            max_new_tokens=max_new_tokens,
            temperature=temperature,
            top_p=top_p,
        )

        a_conv = avg_pairwise_similarity(resp_a)
        b_conv = avg_pairwise_similarity(resp_b)
        cross_sim = cross_similarity(resp_a, resp_b)

        a_conv_scores.append(a_conv)
        b_conv_scores.append(b_conv)
        cross_scores.append(cross_sim)

        all_results.append({
            "query": q,
            "full_sft_responses": resp_a,
            "change_selective_responses": resp_b,
            "full_sft_intra_convergence": a_conv,
            "change_selective_intra_convergence": b_conv,
            "cross_model_similarity": cross_sim,
        })

    metrics = {
        "full_sft_avg_intra_convergence": float(np.mean(a_conv_scores)),
        "change_selective_avg_intra_convergence": float(np.mean(b_conv_scores)),
        "avg_cross_model_similarity": float(np.mean(cross_scores)),
    }
    return metrics, all_results
```

```python
# ============================================================
# 12. CHANGE-BASED NEURON SCORING
# ============================================================

def compute_changed_neuron_scores(
    init_model,
    full_model,
) -> Dict[int, torch.Tensor]:
    """
    각 layer의 각 뉴런에 대해 변화량 점수 계산
    점수:
      ||delta gate[n,:]||_2
    + ||delta up[n,:]||_2
    + ||delta down[:,n]||_2
    """
    init_layers = init_model.model.layers
    full_layers = full_model.model.layers

    layer_scores = {}

    for layer_idx, (init_layer, full_layer) in enumerate(zip(init_layers, full_layers)):
        init_mlp = init_layer.mlp
        full_mlp = full_layer.mlp

        delta_gate = (full_mlp.gate_proj.weight.detach().float().cpu() -
                      init_mlp.gate_proj.weight.detach().float().cpu())
        delta_up = (full_mlp.up_proj.weight.detach().float().cpu() -
                    init_mlp.up_proj.weight.detach().float().cpu())
        delta_down = (full_mlp.down_proj.weight.detach().float().cpu() -
                      init_mlp.down_proj.weight.detach().float().cpu())

        gate_norm = torch.norm(delta_gate, dim=1)        # [intermediate_size]
        up_norm = torch.norm(delta_up, dim=1)            # [intermediate_size]
        down_norm = torch.norm(delta_down, dim=0)        # [intermediate_size]

        score = gate_norm + up_norm + down_norm
        layer_scores[layer_idx] = score

    return layer_scores


def select_top_changed_neurons(
    layer_scores: Dict[int, torch.Tensor],
    top_ratio: float,
):
    flat = []
    for layer_idx, score_vec in layer_scores.items():
        for neuron_idx, score in enumerate(score_vec.tolist()):
            flat.append((layer_idx, neuron_idx, float(score)))

    flat.sort(key=lambda x: x[2], reverse=True)
    k = max(1, int(len(flat) * top_ratio))
    selected = flat[:k]

    selected_map: Dict[int, set] = {}
    for layer_idx, neuron_idx, _ in selected:
        selected_map.setdefault(layer_idx, set()).add(neuron_idx)

    return selected_map, selected
```

```python
# ============================================================
# 13. QWEN NEURON GRADIENT MASKER
# ============================================================

class QwenNeuronGradientMasker:
    """
    선택된 뉴런 n에 대응:
      gate_proj.weight[n, :]
      up_proj.weight[n, :]
      down_proj.weight[:, n]
    만 gradient 유지
    """

    def __init__(self, model: nn.Module, selected_map: Dict[int, set]):
        self.model = model
        self.selected_map = selected_map
        self.handles = []

    def register(self):
        self.remove()

        for layer_idx, layer in enumerate(self.model.model.layers):
            selected = self.selected_map.get(layer_idx, set())
            mlp = layer.mlp

            gate_w_mask = torch.zeros_like(mlp.gate_proj.weight)
            up_w_mask = torch.zeros_like(mlp.up_proj.weight)
            down_w_mask = torch.zeros_like(mlp.down_proj.weight)

            gate_b_mask = torch.zeros_like(mlp.gate_proj.bias) if mlp.gate_proj.bias is not None else None
            up_b_mask = torch.zeros_like(mlp.up_proj.bias) if mlp.up_proj.bias is not None else None

            for n in selected:
                gate_w_mask[n, :] = 1.0
                up_w_mask[n, :] = 1.0
                down_w_mask[:, n] = 1.0

                if gate_b_mask is not None:
                    gate_b_mask[n] = 1.0
                if up_b_mask is not None:
                    up_b_mask[n] = 1.0

            self.handles.append(
                mlp.gate_proj.weight.register_hook(lambda g, m=gate_w_mask: g * m.to(g.device))
            )
            self.handles.append(
                mlp.up_proj.weight.register_hook(lambda g, m=up_w_mask: g * m.to(g.device))
            )
            self.handles.append(
                mlp.down_proj.weight.register_hook(lambda g, m=down_w_mask: g * m.to(g.device))
            )

            if mlp.gate_proj.bias is not None:
                self.handles.append(
                    mlp.gate_proj.bias.register_hook(lambda g, m=gate_b_mask: g * m.to(g.device))
                )
            if mlp.up_proj.bias is not None:
                self.handles.append(
                    mlp.up_proj.bias.register_hook(lambda g, m=up_b_mask: g * m.to(g.device))
                )

            if mlp.down_proj.bias is not None:
                self.handles.append(
                    mlp.down_proj.bias.register_hook(lambda g: torch.zeros_like(g))
                )

    def remove(self):
        for h in self.handles:
            h.remove()
        self.handles = []
```

```python
# ============================================================
# 14. TRAINING LOOP (WITH PROGRESS DISPLAY)
# ============================================================

def train_one_experiment(
    exp_name: str,
    model,
    tokenizer,
    train_loader,
    val_loader,
    cfg: Config,
    device: str,
    selected_map: Dict[int, set] = None,   # None이면 full SFT
):
    exp_dir = os.path.join(cfg.output_dir, exp_name)
    ensure_dir(exp_dir)

    optimizer = torch.optim.AdamW(
        model.parameters(),
        lr=cfg.learning_rate,
        weight_decay=cfg.weight_decay,
    )

    masker = None
    if selected_map is not None:
        masker = QwenNeuronGradientMasker(model, selected_map)
        masker.register()

    global_step = 0
    optimizer_step = 0
    log_history = []

    total_steps_per_epoch = len(train_loader)
    total_global_steps = total_steps_per_epoch * cfg.num_epochs

    print(f"\n===== START TRAINING: {exp_name} =====")
    print(f"total_steps_per_epoch = {total_steps_per_epoch}")
    print(f"total_global_steps    = {total_global_steps}")
    print(f"grad_accum_steps      = {cfg.grad_accum_steps}")
    print(f"selected_neuron_count = {0 if selected_map is None else sum(len(v) for v in selected_map.values())}")

    for epoch in range(cfg.num_epochs):
        for step_in_epoch, batch in enumerate(train_loader, start=1):
            global_step += 1
            batch = {k: v.to(device) for k, v in batch.items()}

            outputs = model(**batch)
            loss = outputs.loss / cfg.grad_accum_steps
            loss.backward()

            did_optimizer_step = False

            if global_step % cfg.grad_accum_steps == 0:
                torch.nn.utils.clip_grad_norm_(model.parameters(), cfg.max_grad_norm)
                optimizer.step()
                optimizer.zero_grad()
                optimizer_step += 1
                did_optimizer_step = True

            # --------------------------
            # train progress log
            # --------------------------
            if (global_step % cfg.log_every == 0) or (global_step == 1) or (global_step == total_global_steps):
                current_loss = float(loss.item() * cfg.grad_accum_steps)

                msg = (
                    f"[{exp_name}] "
                    f"epoch {epoch+1}/{cfg.num_epochs} | "
                    f"step {step_in_epoch}/{total_steps_per_epoch} | "
                    f"global {global_step}/{total_global_steps} | "
                    f"opt_step {optimizer_step} | "
                    f"train_loss {current_loss:.4f}"
                )

                if did_optimizer_step:
                    msg += " | optimizer_step=YES"
                else:
                    msg += " | optimizer_step=NO"

                print(msg)

                row = {
                    "epoch": epoch + 1,
                    "step_in_epoch": step_in_epoch,
                    "global_step": global_step,
                    "optimizer_step": optimizer_step,
                    "train_loss": current_loss,
                }
                log_history.append(row)

            # --------------------------
            # val loss only evaluation
            # --------------------------
            if (global_step % cfg.eval_every_loss_only == 0) or (global_step == total_global_steps):
                loss_metrics = evaluate_loss(model, val_loader, device)

                msg = (
                    f"[{exp_name}] "
                    f"VALIDATION @ global {global_step}/{total_global_steps} | "
                    f"val_loss {loss_metrics['val_loss']:.4f} | "
                    f"val_ppl {loss_metrics['val_ppl']:.2f}"
                )
                print(msg)

                row = {
                    "epoch": epoch + 1,
                    "step_in_epoch": step_in_epoch,
                    "global_step": global_step,
                    "optimizer_step": optimizer_step,
                    **loss_metrics,
                }
                log_history.append(row)
                save_json(log_history, os.path.join(exp_dir, "log_history.json"))

    final_loss_metrics = evaluate_loss(model, val_loader, device)
    final_result = {
        "exp_name": exp_name,
        **final_loss_metrics,
        "selected_neuron_count": 0 if selected_map is None else sum(len(v) for v in selected_map.values()),
        "total_global_steps": total_global_steps,
        "total_optimizer_steps": optimizer_step,
    }

    save_json(final_result, os.path.join(exp_dir, "final_result.json"))
    save_json(log_history, os.path.join(exp_dir, "log_history.json"))

    model.save_pretrained(os.path.join(exp_dir, "model"))
    tokenizer.save_pretrained(os.path.join(exp_dir, "model"))

    if masker is not None:
        masker.remove()

    print(f"\n===== FINISHED TRAINING: {exp_name} =====")
    print(json.dumps(final_result, ensure_ascii=False, indent=2))

    return final_result
```

```python
# ============================================================
# 15. RUN FULL SFT
# ============================================================

save_json(asdict(config), os.path.join(config.output_dir, "config.json"))

print("\n===== Running Full SFT =====")
model_full, tokenizer_full = build_model_and_tokenizer(config, device)

result_full = train_one_experiment(
    exp_name="full_sft",
    model=model_full,
    tokenizer=tokenizer_full,
    train_loader=train_loader,
    val_loader=val_loader,
    cfg=config,
    device=device,
    selected_map=None,
)

print("FULL SFT RESULT:")
print(json.dumps(result_full, ensure_ascii=False, indent=2))
```
